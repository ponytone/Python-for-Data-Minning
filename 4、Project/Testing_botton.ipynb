```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

import sys
import pickle
import re
import codecs
import string
import shutil
from win32com import client as wc
import docx
import os
import time

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
```

```python
# funtion set

def setting_func(num):
#     from selenium import webdriver
#     driver = webdriver.Chrome()
#     driver.get('https://www.google.com/')
#     driver.maximize_window()

    from pynput.mouse import Controller,Button
    mouse = Controller()

    mouse.position = (1420,40)
    time.sleep(0.5)

    mouse.press(Button.left)
    mouse.release(Button.left)

    mouse.position = (1300,410)
    time.sleep(0.5)

    mouse.press(Button.left)
    mouse.release(Button.left)

    time.sleep(1)

    from pynput.keyboard import Controller,Key
    keyboard = Controller()
    keyboard.type('下載')

    time.sleep(1)
    mouse.scroll(0,-500)

    mouse.position = (1000,700)

    time.sleep(1)
    mouse.press(Button.left)
    mouse.release(Button.left)

    time.sleep(1)
    keyboard.press(Key.shift)
    keyboard.release(Key.shift)
    keyboard.type(num)
    keyboard.press(Key.enter)
    keyboard.release(Key.enter)

    time.sleep(1)
    keyboard.press(Key.enter)
    keyboard.release(Key.enter)
    
# download move
def download_move():
    from pynput.keyboard import Controller,Key
    keyboard = Controller()
    with keyboard.pressed(Key.ctrl):
        with keyboard.pressed(Key.shift):
            keyboard.press('j')
            keyboard.release('j')

# last page
def last_page():
    try:
        select = Select(driver.find_element_by_name('pag'))
        select.select_by_index(len(driver.find_elements_by_tag_name('select')[2].find_elements_by_tag_name('option')) - 1)
    except:
        pass
    
# close page
def close_page():
    from pynput.keyboard import Controller,Key
    keyboard = Controller()
    with keyboard.pressed(Key.ctrl):
        keyboard.press('w')
        keyboard.release('w')
        
# phase on
def phase_on():
    from pynput.keyboard import Controller,Key
    keyboard = Controller()
    with keyboard.pressed(Key.ctrl):
        keyboard.press('v')
        keyboard.release('v')
    keyboard.press(Key.enter)
    keyboard.release(Key.enter)
    
# multiple files
def multiple_files():
    from pynput.mouse import Controller,Button
    mouse = Controller()
    mouse.position = (325,160)
    mouse.press(Button.left)
    mouse.release(Button.left)
```
